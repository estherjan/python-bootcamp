```python
#!/usr/bin/env python
# HW02_ch03_ex03

# This exercise can be done using only the statements and other features we
# have learned so far.

# (1) Write a function that draws a grid like the following:
# + - - - - + - - - - +
# |         |         |
# |         |         |
# |         |         |
# |         |         |
# + - - - - + - - - - +
# |         |         |
# |         |         |
# |         |         |
# |         |         |
# + - - - - + - - - - +

# Hint: to print more than one value on a line, you can print a
# comma-separated sequence of values:

# print('+', '-')

# By default, print advances to the next line, but you can
# override that behavior and put a space at the end, like this:

# print('+', end=' ')
# print('-')

# The output of these statements is '+ -'.

# A print statement with no argument ends the current line and
# goes to the next line.
def two_by_two():
    def dividing_line():
        print('+', end = ' ')
        print(('- ') * 4, end = '')
        print('+', end = ' ')
        print(('- ') * 4, end = '')
        print('+')

    def body_line():
        print('|', end = ' ')
        print('  ' * 4, end = '')
        print('|', end = ' ')
        print('  ' * 4, end = '')
        print('|')

    def print_twice(f):
        f()
        f()

    def one_row():
        dividing_line()
        print_twice(body_line)
        print_twice(body_line)

    print_twice(one_row)
    dividing_line()

two_by_two()


# (2) Write a function that draws a similar grid with four rows and four
# columns.
###############################################################################
# Write your functions below:
# Body
def four_by_four():
    def dividing_line():
        print('+', end = ' ')
        print(('- ') * 4, end = '')
        print('+', end = ' ')
        print(('- ') * 4, end = '')
        print('+', end = ' ')
        print(('- ') * 4, end = '')
        print('+', end = ' ')
        print(('- ') * 4, end = '')
        print('+')

    def body_line():
        print('|', end = ' ')
        print('  ' * 4, end = '')
        print('|', end = ' ')
        print('  ' * 4, end = '')
        print('|', end = ' ')
        print('  ' * 4, end = '')
        print('|', end = ' ')
        print('  ' * 4, end = '')
        print('|')

    def print_twice(f):
        f()
        f()
    
    def print_four(f):
        print_twice(f)
        print_twice(f)

    def one_row():
        dividing_line()
        print_four(body_line)

    print_four(one_row)
    dividing_line()
    
four_by_four()

# Write your functions above:
###############################################################################
def main():
    """Call your functions within this function.
    When complete have two function calls in this function:
    two_by_two()
    four_by_four()
    """
    two_by_two()
    four_by_four()


if __name__ == "__main__":
    main()
```
