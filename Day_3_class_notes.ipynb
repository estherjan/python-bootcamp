## break & continue

```python
for i in range(0, 50):
    if i % 2 != 0:
        continue
    print(i)
```

```python
for i in range(0, 50):
    if i % 2 != 0:
        pass  #means "keep going".
    print(i)
```

```python
for i in range(0, 50):
    if i % 2 is not 0: #"is not" works the same as "!=".
        continue
    print(i)
```
